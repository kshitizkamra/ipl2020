```python
for index,rows in df_dl.iterrows():
    sys.stdout.write("\r "+ str(index) + "of " + str(df_dl.shape[0]))
    sys.stdout.flush()
    check_location=str(rows.match_id)+rows.batsman
    try:
        player_performance.index.get_loc(check_location)
    except:
        player_performance.loc[check_location,'player']=rows.batsman
        player_performance.loc[check_location,'match_id']=rows.match_id
        player_performance.loc[check_location,'against']=rows.bowling_team
        player_performance.fillna(0,inplace=True) 
    
    if player_performance.loc[check_location,'inning']==0 :
        player_performance.loc[check_location,'inning']=rows.inning
    if df_mt['team1'][df_mt['id']==rows.match_id].values==rows.batting_team:
        player_performance.loc[check_location,'home']=1
        
    if df_mt['toss_winner'][df_mt['id']==rows.match_id].values==rows.batting_team:
        player_performance.loc[check_location,'toss_winner']=1    
    player_performance.loc[check_location,'season']=df_mt['season'][df_mt['id']==rows.match_id].values
    player_performance.loc[check_location,'runs']=player_performance.loc[check_location,'runs']+rows.batsman_runs
    if rows.wide_runs==0:
        player_performance.loc[check_location,'balls_faced']=player_performance.loc[check_location,'balls_faced']+1
        if player_performance.loc[check_location,'balls_faced']>0:
            player_performance.loc[check_location,'strike_rate']=player_performance.loc[check_location,'runs']/player_performance.loc[check_location,'balls_faced']*100
    if rows.batsman_runs==4:
        player_performance.loc[check_location,'fours']=player_performance.loc[check_location,'fours']+1
        player_performance.loc[check_location,'points']=player_performance.loc[check_location,'points']+1
    if rows.batsman_runs==6:
        player_performance.loc[check_location,'sixes']=player_performance.loc[check_location,'sixes']+1
        player_performance.loc[check_location,'points']=player_performance.loc[check_location,'points']+2
    if rows.player_dismissed==player_performance.loc[check_location,'player']:
        player_performance.loc[check_location,'dismissed']=1
        if player_performance.loc[check_location,'runs']==0:
            player_performance.loc[check_location,'points']=player_performance.loc[check_location,'points']-2
    player_performance.loc[check_location,'points']=player_performance.loc[check_location,'points']+rows.batsman_runs
    if player_performance.loc[check_location,'runs']>=100:
        player_performance.loc[check_location,'centuary_points']=16
    elif player_performance.loc[check_location,'runs']>=50:
        player_performance.loc[check_location,'centuary_points']=8
    if player_performance.loc[check_location,'balls_faced']>=10:
        if player_performance.loc[check_location,'strike_rate']<50:
            player_performance.loc[check_location,'sr_points']=-6
        elif player_performance.loc[check_location,'strike_rate']<60:
            player_performance.loc[check_location,'sr_points']=-4
        elif player_performance.loc[check_location,'strike_rate']<70:
            player_performance.loc[check_location,'sr_points']=-2
        else :
             player_performance.loc[check_location,'sr_points']=0
        
    
    
        

        
    check_location_bowler=str(rows.match_id)+rows.bowler
    try:
        player_performance.index.get_loc(check_location_bowler)
    except:
        player_performance.loc[check_location_bowler,'player']=rows.bowler
        player_performance.loc[check_location_bowler,'match_id']=rows.match_id
        player_performance.loc[check_location_bowler,'against']=rows.batting_team
        player_performance.fillna(0,inplace=True)
    
    if df_mt['team1'][df_mt['id']==rows.match_id].values==rows.bowling_team:
        player_performance.loc[check_location_bowler,'home']=1
        
    player_performance.loc[check_location_bowler,'season']=df_mt['season'][df_mt['id']==rows.match_id].values
    if df_mt['toss_winner'][df_mt['id']==rows.match_id].values==rows.bowling_team:
        player_performance.loc[check_location_bowler,'toss_winner']=1
        
    
    if player_performance.loc[check_location_bowler,'inning']==0:
        if rows.inning==1:
            player_performance.loc[check_location_bowler,'inning']=2
        else :
            player_performance.loc[check_location_bowler,'inning']=1
            
    
    if rows.noball_runs==0 and rows.wide_runs==0:
        player_performance.loc[check_location_bowler,'balls_bowled']=player_performance.loc[check_location_bowler,'balls_bowled']+1
    player_performance.loc[check_location_bowler,'runs_given']=player_performance.loc[check_location_bowler,'runs_given']+rows.total_runs-rows.legbye_runs-rows.bye_runs-rows.penalty_runs
    if rows.dismissal_kind!='run out' and rows.dismissal_kind!=0 and rows.dismissal_kind!='retired hurt' and rows.dismissal_kind!='obstructing the field' and rows.dismissal_kind!='0':
        player_performance.loc[check_location_bowler,'wickets']=player_performance.loc[check_location_bowler,'wickets']+1
        player_performance.loc[check_location_bowler,'points']=player_performance.loc[check_location_bowler,'points']+25
        if player_performance.loc[check_location_bowler,'wickets']==5:
            player_performance.loc[check_location_bowler,'points']=player_performance.loc[check_location_bowler,'points']+8
        elif player_performance.loc[check_location_bowler,'wickets']==4:
            player_performance.loc[check_location_bowler,'points']=player_performance.loc[check_location_bowler,'points']+8
    if player_performance.loc[check_location_bowler,'balls_bowled']>0:
        overs=int(player_performance.loc[check_location_bowler,'balls_bowled']/6)
        balls=player_performance.loc[check_location_bowler,'balls_bowled']-(overs*6)
        overs=overs+balls/6
        player_performance.loc[check_location_bowler,'economy_rate']=player_performance.loc[check_location_bowler,'runs_given']/overs
    if player_performance.loc[check_location_bowler,'balls_bowled']>=12:
        if player_performance.loc[check_location_bowler,'economy_rate']<4:
            player_performance.loc[check_location_bowler,'er_points']=6
        elif player_performance.loc[check_location_bowler,'economy_rate']<5:
            player_performance.loc[check_location_bowler,'er_points']=4
        elif player_performance.loc[check_location_bowler,'economy_rate']<6:
            player_performance.loc[check_location_bowler,'er_points']=2
        elif player_performance.loc[check_location_bowler,'economy_rate']>=9 and player_performance.loc[check_location_bowler,'economy_rate']<=10:
            player_performance.loc[check_location_bowler,'er_points']=-2
        elif player_performance.loc[check_location_bowler,'economy_rate']>10 and player_performance.loc[check_location_bowler,'economy_rate']<=11:
            player_performance.loc[check_location_bowler,'er_points']=-4
        elif player_performance.loc[check_location_bowler,'economy_rate']>11:
            player_performance.loc[check_location_bowler,'er_points']=-6
        else:
            player_performance.loc[check_location_bowler,'er_points']=0
        


    if rows.fielder!=0:
        check_location_fielder=str(rows.match_id)+rows.fielder
        try:
            player_performance.index.get_loc(check_location_fielder)
            
        except:
            player_performance.loc[check_location_fielder,'player']=rows.fielder
            player_performance.loc[check_location_fielder,'against']=rows.batting_team
            player_performance.fillna(0,inplace=True)
            player_performance.loc[check_location_fielder,'match_id']=rows.match_id
            
        if df_mt['team1'][df_mt['id']==rows.match_id].values==rows.bowling_team:
            player_performance.loc[check_location_fielder,'home']=1
    
        if df_mt['toss_winner'][df_mt['id']==rows.match_id].values==rows.bowling_team:
            player_performance.loc[check_location_fielder,'toss_winner']=1    
            
        player_performance.loc[check_location_fielder,'season']=df_mt['season'][df_mt['id']==rows.match_id].values
        if rows.dismissal_kind=='caught' or rows.dismissal_kind=='caught and bowled' :
            player_performance.loc[check_location_fielder,'catches']=player_performance.loc[check_location_fielder,'catches']+1
            player_performance.loc[check_location_fielder,'points']=player_performance.loc[check_location_fielder,'points']+8
        if rows.dismissal_kind=='run out':
            player_performance.loc[check_location_fielder,'runout_throw']=player_performance.loc[check_location_fielder,'runout_throw']+1
            player_performance.loc[check_location_fielder,'points']=player_performance.loc[check_location_fielder,'points']+6
        if rows.dismissal_kind=='stumped':
            player_performance.loc[check_location_fielder,'stumpings']=player_performance.loc[check_location_fielder,'stumpings']+1
            player_performance.loc[check_location_fielder,'points']=player_performance.loc[check_location_fielder,'points']+12
            
    
    
```
